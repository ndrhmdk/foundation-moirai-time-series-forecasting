```python
import sys
import os
import matplotlib.pyplot as plt
import yaml
import pandas as pd
```

```python
sys.path.append(os.path.abspath(os.path.join('..')))
from src.dataloader import load_data, get_splits
```

```python
with open("../configs/config.yaml", 'r') as f:
    config = yaml.safe_load(f)
```

```python
def visualize(name, df):
    plt.figure(figsize=(12, 4))
    plt.plot(df.index, df['values'])
    plt.title(f"{name}")
    plt.grid(True, linestyle='--', alpha=0.4)
    plt.show()
```

```python
df_spy = load_data("SPY", config)
print(f"SPY Range: {df_spy.index.min()} to {df_spy.index.max()}")
visualize("SPY", df_spy)
```

## **Save to CSV**

```python
DATASETS_DIR = f"../datasets"
SPY_CSV = os.path.join(DATASETS_DIR, "SPY.csv")
AQI_CSV = os.path.join(DATASETS_DIR, "AQI.csv")
```

```python
df_spy.to_csv(SPY_CSV)
temp_df = pd.read_csv(f"{SPY_CSV}", index_col='date', parse_dates=True)
print(f"SPY Range: {temp_df.index.min()} to {temp_df.index.max()}")
visualize("SPY", temp_df)
```

```python
df_aqi = load_data("AQI_LA", config)
print(f"AQI Range: {df_aqi.index.min()} to {df_aqi.index.max()}")
visualize("AQI", df_aqi)
```

```python
df_aqi.to_csv(AQI_CSV)
temp_df = pd.read_csv(f"{AQI_CSV}", index_col='date', parse_dates=True)
print(f"AQI Range: {temp_df.index.min()} to {temp_df.index.max()}")
visualize("AQI", temp_df)
```
